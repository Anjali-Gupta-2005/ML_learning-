```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()

X= iris.data
y= iris.target
```

```python
#scale 

scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
# visualise- abhi khali do hi feature mei kar rahe hai
sns.scatterplot(x= X_scaled[:,0], y=X[:,2])
```

```python
# plotting the dendogram to find no of clusters

from scipy.cluster.hierarchy import linkage, dendrogram

#linkage matrix
#to store dendogram matrix
Z= linkage(X_scaled, method="ward")

#plot

plt.figure(figsize=(2,6))
dendrogram(Z)
plt.xlabel("samples")
plt.ylabel("distance")
plt.title("dendrogram for hirarchical clustering")
```

```python
# clustering for k=2 even though iris set has 3
from sklearn.cluster import AgglomerativeClustering 

agg= AgglomerativeClustering(n_clusters=2)
labels=agg.fit_predict(X_scaled)
```

```python
sns.scatterplot(x= X_scaled[:,0], y=X[:,2],c=labels)
```
