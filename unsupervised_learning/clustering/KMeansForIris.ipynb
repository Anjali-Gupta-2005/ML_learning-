```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
```

```python
iris= load_iris()
```

```python
X= iris.data
y=iris.target
```

```python
X.shape
```

```python
# Visualize 
# randomly 2 featurs hum visualize kar rahe hai but its not accurate though
```

```python
sns.scatterplot(x=X[:,0],y=X[:,2])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
# optional - dimensionality reduction using PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 4 features se 2 karliye
pca_data= pca.fit_transform(X_scaled)
```

```python
# Elbow method 

wcss= []

for k in range(1,11):
    kmeans= KMeans(n_clusters=k)
    kmeans.fit(pca_data)
    wcss.append(kmeans.inertia_)
    
```

```python
sns.lineplot(x= range(1,11),y=wcss, marker='o')
```

```python
# kMeans

kmeans= KMeans(n_clusters=3,random_state=10)
labels= kmeans.fit_predict(pca_data)

sns.scatterplot(x=pca_data[:,0],y=pca_data[:,1],c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y=kmeans.cluster_centers_[:,1],marker="x",c="red",s=200)
```
