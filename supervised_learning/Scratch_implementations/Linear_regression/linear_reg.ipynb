```python
import numpy as np
```

```python
#linear Regression

class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iter=1000):
        self.bias= None
        self.weights= None
        self.lr= learning_rate
        self.n_iter = n_iter
    def fit(self,X,y): #X_train, y_train
      m,n=  X.shape # tells no of samples, no of features
      # random initialisation
      self.bias=0
      self.weights= np.zeros(n) # return array with zeros

    # Gradient Descent
      for i in range(self.n_iter):
        # calculate y pred 
        y_pred= self.bias+ np.dot(X,self.weights)

        # calculating derivatives/ gradient of cost function
        db = (1/m)*np.sum(y_pred-y) # derivative with respect to bias
        dw = (1/m)*np.dot(X.T,(y_pred-y)) # derivative with respect to weights

        # convergence theorem- params update
        self.bias-= self.lr*db
        self.weights-= self.lr*dw
      
    def predict(self, X):
        y_pred=self.bias+ np.dot(X,self.weights)
        return y_pred
        
```

```python
X= np.array([[1],[2],[3],[4],[5]])
y=np.array([2,4,6,8,10])

model = LinearRegression()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)
```

```python
#Linear reg with OLS

class LinearRegressionOLS:
    def __init__(self):
        self.bias= None
        self.weights= None
    def fit(self,X,y):
         m,n=  X.shape
         #to add a column with all ones
         X_b=np.c_[np.ones((m,1)),X ]
        
         thetha = np.dot(np.linalg.inv(np.dot(X_b.T,X_b)),np.dot(X_b.T,y))
        #instead of np.dot() we can write A@B
         self.weights=thetha[1:]
         self.bias=thetha[0]
    def predict(self,X):
        y_pred= self.bias+ np.dot(X,self.weights)
        return y_pred
```

```python
X= np.array([[1],[2],[3],[4],[5]])
y=np.array([2,4,6,8,10])

model = LinearRegressionOLS()
model.fit(X,y)

y_pred = model.predict(X)
print(y_pred)
```
